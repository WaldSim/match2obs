```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# daten laden
df20 = pd.read_csv("LUBW/HD_2020.csv", usecols = ["Datum ", "Temperatur (°C)", "NO (µg/m³)", "NO2 (µg/m³)", "Windrichtung (°)"], encoding="ISO-8859-1", infer_datetime_format =True, dayfirst = True) #skiprows = [3])
df20 = df20.rename({'NO2 (µg/m³)': 'NO2_20 (µg/m³)', "NO (µg/m³)":"NO_20 (µg/m³)", "Temperatur (°C)":"Temperatur_20 (°C)", "Windrichtung (°)":"Windrichtung_20 (°)" }, axis=1)
df19 = pd.read_csv("LUBW/HD_2019.csv", usecols = ["Datum ", "Temperatur (°C)", "NO (µg/m³)", "NO2 (µg/m³)"],encoding="ISO-8859-1", infer_datetime_format =True, dayfirst = True) #skiprows = [3])
df19 = df19.rename({'NO2 (µg/m³)': 'NO2_19 (µg/m³)', "NO (µg/m³)":"NO_19 (µg/m³)", "Temperatur (°C)":"Temperatur_19 (°C)", "Windrichtung (°)":"Windrichtung_19 (°)" }, axis=1)
df18 = pd.read_csv("LUBW/HD_2018.csv", usecols = ["Datum ", "Temperatur (°C)", "NO (µg/m³)", "NO2 (µg/m³)", "Windrichtung (°)"], encoding="ISO-8859-1", infer_datetime_format =True, dayfirst = True) #skiprows = [3])
df18 = df18.rename({'NO2 (µg/m³)': 'NO2_18 (µg/m³)', "NO (µg/m³)":"NO_18 (µg/m³)", "Temperatur (°C)":"Temperatur_18 (°C)", "Windrichtung (°)":"Windrichtung_18 (°)" }, axis=1)
df17 = pd.read_csv("LUBW/HD_2017.csv", usecols = ["Datum ", "Temperatur (°C)", "NO (µg/m³)", "NO2 (µg/m³)", "Windrichtung (°)"], encoding="ISO-8859-1", infer_datetime_format =True, dayfirst = True) #skiprows = [3])
df17 = df17.rename({'NO2 (µg/m³)': 'NO2_17 (µg/m³)', "NO (µg/m³)":"NO_17 (µg/m³)", "Temperatur (°C)":"Temperatur_17 (°C)", "Windrichtung (°)":"Windrichtung_17 (°)" }, axis=1)


# rolling mean über 12 h
df20_roll =df20.rolling(24, on ="Datum ").mean()
df19_roll =df19.rolling(24, on ="Datum ").mean()
df18_roll =df18.rolling(24, on ="Datum ").mean()
df17_roll =df17.rolling(24, on ="Datum ").mean()
```
